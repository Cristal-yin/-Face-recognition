## 简单的线条
matplotlib的figure就是一个单独的figure小窗口， 小窗口里面还可以有更多的小图片。


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-3, 3, 50)
y1 = 2*x + 1
y2 = x**2
```

```python
plt.figure()
plt.plot(x, y1)
plt.show()
```

```python
plt.figure()
plt.plot(x, y2)
plt.show()
```

```python
plt.figure(num = 3, figsize = (8, 5),)
plt.plot(x, y2)
plt.plot(x, y1, color = 'red', linewidth = 1.0, linestyle = '--')
plt.show()
```
